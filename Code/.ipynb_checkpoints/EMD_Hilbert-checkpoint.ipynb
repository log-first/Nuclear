```python
from os import listdir
from scipy import io as scio

import matplotlib.pyplot as plt

import numpy as np 
import pandas as pd 
import pywt
```

```python
fileName = listdir('../data/512S/')
fileName
```

```python
dataCo60_01 = scio.loadmat('../data/512S/Co60_01.mat')
dataCo60_01
```

```python
dataCo60_01 = dataCo60_01['AI_Ch00'][0]
```

```python
len(dataCo60_01)
```

```python
### 绘图
fig, axes = plt.subplots(ncols=2, figsize=(16, 8))

axes[0].plot(dataCo60_01[0 : 512])
axes[1].plot(dataCo60_01[512 * 100 : 512 * 101])

plt.show()
```

```python
a = np.delete(dataCo60_01,np.where(dataCo60_01[:] > -100)[0], axis=0)
```

```python
a, len(a) 
```

```python
data = []
for i in range(len(a)):
    if i > 2:
        if (a[i-1] - a[i - 2] < 0) and (a[i] - a[i-1] < 0) and (a[i+1] - a[i] > 0) and (a[i + 2] - a[i + 1] > 0):
            data.append(a[i])
data = np.array(data)
data = data[:] * -1
```

```python
len(data)
```

```python
data = np.delete(data, np.where(data[:] > 6000)[0], axis=0)
```

```python
plt.hist(data, bins=300, normed=0, facecolor="blue", edgecolor="black", alpha=0.7)
plt.show()
```

```python
dataCs137 = scio.loadmat('../data/512S/Cs137_01.mat')
dataCs137 = dataCs137['AI_Ch00'][0]
dataCs137
```

```python
data = []
for i in range(len(b)):
    if i > 2:
        if (b[i-1] - b[i - 2] < 0) and (b[i] - b[i-1] < 0) and (b[i+1] - a[i] > 0) and (b[i + 2] - b[i + 1] > 0):
            data.append(b[i])
data = np.array(data)
```
